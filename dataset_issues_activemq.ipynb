```python
import pandas as pd
import os
```

```python
def convert_file_issues_to_list_issues(my_files):
  list_issues = []
  dict_issue = {}
  for filename in my_files:
    with open(f'my_issues/{filename}', 'r') as f:
      content_file = f.read()

    issue_key = filename
    issue_type = content_file.split('issue_type:')[1].split('\n')[0]
    issue_type = issue_type.replace(' ', '')
    summary = content_file.split('summary:')[1].split('\n')[0]
    description = content_file.split('description:')[1].split('status:')[0]
    status = content_file.split('status:')[1].split('\n')[0]
    comments = content_file.split('status:')[1].split('comments:')[1]

    dict_issue = {
          'issue_key': issue_key,
          'issue_type': issue_type,
          'summary': summary,
          'description': description,
          'status': status,
          'comments': comments
    }
    list_issues.append(dict_issue)
  return list_issues
```

```python
my_files = os.listdir('my_issues')
list_issues = convert_file_issues_to_list_issues(my_files=my_files)
df_my_activemq_selected_issues = pd.DataFrame(list_issues)
```

```python
df_activemq_issues_inspected = pd.read_excel('activemq_issues_inspected.xlsx')
df_activemq_issues_inspected
```

```python
# Assuming 'issue_key' is the common column in both dataframes
df_merged = pd.merge(df_activemq_issues_inspected, df_my_activemq_selected_issues, on='issue_key', how='left')
df_merged
```

```python
df_activemq_issues_merged = df_merged[['issue_key', 'issue_type', 'status', 'summary_x',
       'description', 'architectural_impact', 'comments']]

df_activemq_issues_merged.rename(columns={'summary_x':'summary'}, inplace=True)
```
