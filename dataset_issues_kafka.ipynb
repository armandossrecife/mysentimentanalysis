# Importa dependências

```python
import pandas as pd
import os
```

```python
def convert_file_issues_to_list_issues(my_files):
  list_issues = []
  dict_issue = {}
  for filename in my_files:
    with open(f'my_issues/{filename}', 'r') as f:
      content_file = f.read()

    issue_key = filename
    issue_type = content_file.split('issue_type:')[1].split('\n')[0]
    issue_type = issue_type.replace(' ', '')
    summary = content_file.split('summary:')[1].split('\n')[0]
    description = content_file.split('description:')[1].split('status:')[0]
    status = content_file.split('status:')[1].split('\n')[0]
    comments = content_file.split('status:')[1].split('comments:')[1]

    dict_issue = {
          'issue_key': issue_key,
          'issue_type': issue_type,
          'summary': summary,
          'description': description,
          'status': status,
          'comments': comments
    }
    list_issues.append(dict_issue)
  return list_issues
```

# Cria o dataframe com as informações e resultados das inspeções

```python
my_files = os.listdir('my_issues')
list_issues = convert_file_issues_to_list_issues(my_files=my_files)
df_my_kafka_selected_issues = pd.DataFrame(list_issues)
```

```python
df_my_kafka_selected_issues
```

```python
df_kafka_issues_inspected = pd.read_excel('kafka_issues_inspected.xlsx')

# Assuming 'issue_key' is the common column in both dataframes
df_merged = pd.merge(df_kafka_issues_inspected, df_my_kafka_selected_issues, on='issue_key', how='left')

df_kafka_issues_merged = df_merged[['issue_key', 'issue_type_x', 'status_x', 'created_date', 'resolved_date', 'time_resolution',
                                    'summary_x', 'description',  'architectural_impact','comments']]

df_kafka_issues_merged.rename(columns={'issue_type_x': 'issue_type', 'status_x': 'status', 'summary_x':'summary'}, inplace=True)
```
